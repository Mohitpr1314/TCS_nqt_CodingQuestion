# Convert Binary to Decimal
Convert a binary number to a decimal number.

Examples:

Example 1:
Input: N = 1011
Output: 11
Explanation: 1011 when converted to decimal number is “11”.

Example 2:
Input: 100
Output: 4
Explanation: 100 when converted to decimal number is “4”.

```python
def binarytodecimal(n):
    decimal = 0
    power = 0
    while n>0:
        dec = n%10
        decimal += dec*(2**power)
        n = n//10
        power += 1
    return decimal

    

n = 1011
binarytodecimal(n)
```

# Convert Decimal to Binary Number
Convert decimal to binary number.

Examples:

Example 1:
Input: N = 15
Output: 1111
Explanation: 15 in binary is represented as "1111".

Example 2:
Input: 18
Output: 10010
Explanation: 18 in binary is represented as "10010"

```python
def decimaltobinary(n):
    ans = ''
    while n>0:
        rem = n%2
        ans = str(rem) + ans
        n = n//2

    return int(ans)

n =56
decimaltobinary(n)
```

# Convert Decimal to Octal
Given a decimal number, convert it into Octal Number.

Examples:

Example 1:
Input:  17
Output: 21
Explanation: Octal Equivalent of 17 is 21

Example 2:
Input:  45
Output: 55
Explanation: Octal Equivalent of 45 is 55

```python
def decimaltooctal(n):
    ans = ''
    while n>0:
        dec = n%8
        ans = str(dec) + ans
        n = n//8
    return int(ans)

n = 45
decimaltooctal(n)
```

# Convert Octal to Decimal
Given an Octal Number, convert it into a Decimal Number.

Examples:

Example 1:
Input: 345
Output: 229
Explanation: Decimal equivalent of given Octal expressionis 229

Example 2:
Input: 170
Output: 121
Explanation: Decimal equivalent of given Octal expression is121


```python
def octaltodecimal(n):
    decimal = 0
    power = 0
    while n>0:
        dec = n%10
        decimal += dec*(8**power)
        n = n//10
        power += 1

    return decimal

n = 345
octaltodecimal(n)
```
